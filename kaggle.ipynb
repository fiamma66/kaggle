```python
import pandas as pd

data = pd.read_csv("../pollution_us_2000_2016.csv")
data.info()
```

```python
data.isna().sum() / len(data) * 100
```

```python
location = data[["Address", "State", "County", "City"]]
location.info()
```

```python
state = location.groupby("Address")["State"].apply(list)
country = location.groupby("Address")["County"].apply(list)
city = location.groupby("Address")["City"].apply(list)
location = pd.DataFrame({
    "Address": state.index,
    "County": country.values,
    "State": state.values,
    "City": city.values
})
location
```

```python
location["County"] = location["County"].apply(lambda x: x[0])
location["State"] = location["State"].apply(lambda x: x[0])
location["City"] = location["City"].apply(lambda x: x[0])
location.info()
```
